```python
import pandas_ta as ta
hdfc_data["rsi"] = ta.rsi(hdfc_data["close"])
```

```python
hdfc_data['sma_20'] = ta.sma(hdfc_data['close'], length=20)
hdfc_data['ema_20'] = ta.ema(hdfc_data['close'], length=20)
```

```python
hdfc_data['rsi_14'] = ta.rsi(hdfc_data['close'], length=14)
```

```python
import pandas as pd
macd = ta.macd(hdfc_data['close'])
hdfc_data = pd.concat([hdfc_data, macd], axis=1)
```

```python
supertrend = ta.supertrend(hdfc_data['high'], hdfc_data['low'], hdfc_data['close'], length=10, multiplier=3)
hdfc_data = pd.concat([hdfc_data, supertrend], axis=1)
```

```python
hdfc_data['rsi_signal'] = hdfc_data['rsi_14'].apply(
    lambda x: 'long' if x < 30 else ('short' if x > 70 else 'hold')
)
hdfc_data['macd_signal'] = hdfc_data.apply(
    lambda row: 'long' if row['MACD_12_26_9'] > row['MACDs_12_26_9']
    else ('short' if row['MACD_12_26_9'] < row['MACDs_12_26_9'] else 'hold'), axis=1
)
hdfc_data['sma_signal'] = hdfc_data.apply(
    lambda row: 'long' if row['close'] > row['sma_20']
    else ('short' if row['close'] < row['sma_20'] else 'hold'), axis=1
)
hdfc_data['ema_signal'] = hdfc_data.apply(
    lambda row: 'long' if row['close'] > row['ema_20']
    else ('short' if row['close'] < row['ema_20'] else 'hold'), axis=1
)
hdfc_data['supertrend_signal'] = hdfc_data.apply(
    lambda row: 'long' if row['close'] > row['SUPERT_10_3.0']
    else ('short' if row['close'] < row['SUPERT_10_3.0'] else 'hold'), axis=1
)

columns_to_show = ['close', 'rsi_signal', 'macd_signal', 'sma_signal', 'ema_signal', 'supertrend_signal']
print(hdfc_data[columns_to_show].tail(10))
```

```python
reliance_data["rsi"] = ta.rsi(reliance_data["close"])
```

```python
reliance_data['sma_20'] = ta.sma(reliance_data['close'], length=20)
reliance_data['ema_20'] = ta.ema(reliance_data['close'], length=20)
```

```python
df = hdfc_data.copy()
rsi_col = 'rsi_14' if 'rsi_14' in df.columns else 'rsi'

df['position'] = 0
df.loc[(df[rsi_col] < 30), 'position'] = 1
df.loc[(df[rsi_col] > 65), 'position'] = -1


df['position'] = df['position'].replace(to_replace=0, method='ffill')


df['position'] = df['position'].shift(1).fillna(0)

df['returns'] = df['close'].pct_change()
df['strategy_returns'] = df['position'] * df['returns']


df['cumulative_strategy'] = (1 + df['strategy_returns']).cumprod()
df['cumulative_market'] = (1 + df['returns']).cumprod()

final_strategy_return = df['cumulative_strategy'].iloc[-1]
final_market_return = df['cumulative_market'].iloc[-1]


trades = df['position'].diff().fillna(0).abs().sum() / 2
winning_trades = df[df['strategy_returns'] > 0].shape[0]
total_trades = df[df['strategy_returns'] != 0].shape[0]
win_rate = (winning_trades / total_trades * 100) if total_trades > 0 else 0


print(f"Final RSI Strategy Return: {final_strategy_return:.2f}x")
print(f"Final Market Return: {final_market_return:.2f}x")
print(f"Total Trades: {int(trades)}")
print(f"Winning Trade Count: {winning_trades}")
print(f"Win Rate: {win_rate:.2f}%")
```
